#Beginner Task Image to Pencil Sketch with Python

In this project, the first step is to choose an image and then read it in RGB format. After that, the image needs to be converted to grayscale, which will make it appear as a black and white photo. The next step involves inverting the grayscale image, also known as the negative image, which can enhance details. Finally, the pencil sketch can be created by combining the grayscale image with the inverted blurry image. This can be achieved by dividing the grayscale image by the inverted blurry image. Since images are represented as arrays, the division operation will be performed using the divide function from the Python cv2 library.

#Importing Module

```python
import cv2
```

#Reading Image

```python
pic1 = cv2.imread('/content/download1.jpg')
cv2_imshow(pic1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#Converting the Original Image into Grayscale Image

Reading the image

```python
gray_img = cv2.cvtColor(pic1, cv2.COLOR_BGR2GRAY)
```

Displaying the Grayscale Image

```python
cv2_imshow(gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#Inverting the Grayscale Image

```python
inverted_img = 255 - gray_img
```

Displaying Inverted Image

```python
cv2_imshow(gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#Blurring the image by using the Gaussian Function in OpenCV

```python
blur_img = cv2.GaussianBlur(inverted_img, (21, 21), 0)
```

Displaying Blurred Image

```python
cv2_imshow(blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#Inverting Blurred Image and Creating a Pencil Sketch

Inverting Blurred Image

```python
inverted_blur_img = 255 - blur_img
```

Creating a Pencil Sketch by dividing the Grayscale Image by Inverted Blurred Image

```python
sketch = cv2.divide(gray_img, inverted_blur_img, scale=256.0)
```

Displaying Pencil Sketch

```python
cv2_imshow(sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Saving the sketch

```python
cv2.imwrite('sketch.png',sketch)
```
